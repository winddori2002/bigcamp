```python
def mk_avg(df):
    df_clmn = df.columns
    new_ls = []
    for _ in range(1,len(df_clmn)):
        ls = list(df[df_clmn[_]])
        avg_value = (sum(ls)-max(ls)-min(ls)) / (len(ls)-2)
        new_ls.append(avg_value)
    new_row = pd.DataFrame([new_ls], columns = df_clmn[1:])
    result = pd.concat([df,new_row])
    return result
```

```python
col=['18Q1PER','18Q2PER','18Q3PER','18Q4PER','19Q1PER','18Q1PBR','18Q2PBR','18Q3PBR','18Q4PBR','19Q1PBR',\
                                '18Q1ROE','18Q2ROE','18Q3ROE','18Q4ROE','19Q1ROE','18Q1EBIT','18Q2EBIT','18Q3EBIT','18Q4EBIT','19Q1EBIT']
fste=pd.DataFrame(mk_avg(TechElec),columns=col)
fsbt=pd.DataFrame(mk_avg(Beauty),columns=col)
fsfd = pd.DataFrame(mk_avg(Food),columns=col)
fsdc = pd.DataFrame(mk_avg(Deco),columns=col)
fsfs = pd.DataFrame(mk_avg(Fashion),columns=col)
fshl =pd.DataFrame(mk_avg(Homeliving),columns=col)
fsls = pd.DataFrame(mk_avg(Leisure),columns=col)
fssp = pd.DataFrame(mk_avg(Sports),columns=col)
fspt = pd.DataFrame(mk_avg(Pet),columns=col)
```
