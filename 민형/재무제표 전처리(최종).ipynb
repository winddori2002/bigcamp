```python
import pandas as pd


TechElec=pd.read_excel('(1)StocksbyCategories.xlsx',sheet_name='TechElec')
Beauty=pd.read_excel('(1)StocksbyCategories.xlsx',sheet_name='Beauty')
Food=pd.read_excel('(1)StocksbyCategories.xlsx',sheet_name='Food')
Deco=pd.read_excel('(1)StocksbyCategories.xlsx',sheet_name='Deco')
Homeliving=pd.read_excel('(1)StocksbyCategories.xlsx',sheet_name='Homeliving')
Leisure=pd.read_excel('(1)StocksbyCategories.xlsx',sheet_name='Leisure')
Sports=pd.read_excel('(1)StocksbyCategories.xlsx',sheet_name='Sports')
Pet=pd.read_excel('(1)StocksbyCategories.xlsx',sheet_name='Pet')
Fashion=pd.read_excel('(1)StocksbyCategories.xlsx',sheet_name='Fashion')

Fashion.head()

categories=[TechElec,Beauty,Food,Deco,Homeliving,Leisure,Sports,Pet,Fashion]

TechElec1=TechElec
Beauty1=Beauty
Food1=Food
Deco1=Deco
Homeliving1=Homeliving
Leisure1=Leisure
Sports1=Sports
Pet1=Pet
Fashion1=Fashion
```

```python
#종목당 한 분기간 지표 값 차이 구하기(4개 모든 지표)
num=[1,2,3]
ind=['PER','PBR','ROE']

#18Q2Q1PER,18Q3Q2PER ~ 19Q1Q4PER(%)  : ex) 18Q2Q1은 18년도 2분기-1분기 차이를 뜻한다, 19Q1Q4은 19년도 1분기-18년도 4분기를 뜻한다.

#18Q2Q1PBR,18Q3Q2PBR ~ 19Q1Q4PBR(%)

#18Q2Q1ROE,18Q3Q2ROE ~ 19Q1Q4ROE(%)

#18Q2Q1EBIT,18Q3Q2EBIT ~ 19Q1Q4EBIT : (후 - 전)/전 *100 (%) 

#18년도 PER,PBR,ROE 분기별 차이 구하기
for i in categories:
    for k in ind:
        for j in num:
             i['18Q{0}Q{1}{2}'.format(j+1,j,k)] = i['18Q{0}{1}'.format(j+1,k)] - i['18Q{0}{1}'.format(j,k)]
                      

     #18년도 EBIT 분기별 차이 구하기
for i in categories:
    for j in num:
        i['18Q{0}Q{1}EBIT'.format(j+1,j)]=(i['18Q{0}EBIT'.format(j+1)]-i['18Q{0}EBIT'.format(j)])*100/i['18Q{0}EBIT'.format(j)]
    
    #19년도 PER,PBR,ROE 분기별 차이 구하기
for i in categories:
    for k in ind:
        i['19Q1Q4{}'.format(k)]=i['19Q1{}'.format(k)]-i['18Q4{}'.format(k)]
       
    #19년도 EBIT 분기별 차이 구하기
for i in categories:
    for j in num:
        i['19Q1Q4EBIT'.format(j+1,j)]=(i['19Q1EBIT'.format(j+1)]-i['18Q4EBIT'.format(j)])*100/i['18Q4EBIT'.format(j)]
```

```python
#전처리 데이터에 펀딩시작날짜 월별로 분기 라벨 부여, 새로운 컬럼 생성
Q119=total[total['펀딩시작날짜']>='2019-04-01']
Q119['분기']='19Q1적용'

Q418=total[(total['펀딩시작날짜']>='2019-01-01')&(total['펀딩시작날짜']<='2019-03-31')]
Q418['분기']='18Q4적용'

Q318=total[(total['펀딩시작날짜']>='2018-10-01')&(total['펀딩시작날짜']<='2018-12-31')]
Q318['분기']='18Q3적용'

Q218=total[(total['펀딩시작날짜']>='2018-01-01')&(total['펀딩시작날짜']<='2018-09-30')]
Q218['분기']='18Q2적용'
```

```python
total_addQ=pd.concat([Q119,Q418,Q318,Q218],axis=0)
total_addQc=total_addQ

total_addQc.columns
total_addQc.head()
```

```python
# for 문 이용하여 시작시기, 카테고리를 매치시켜 4개지표의 차이값을 연결
# (카테고리, 분기)조합별로 새로 객체 생성하여 매칭되는 증감값 컬럼4개 추가하고 마지막에 모두 병합하는 방식

# for문을 위한 사전 리스트 생성
카테고리=['테크·가전', '뷰티', '푸드', '디자인소품', '홈리빙','여행·레저','스포츠·모빌리티','반려동물','패션·잡화']
분기=['18Q2적용','18Q3적용','18Q4적용','19Q1적용']
li1=[]
li2=[]
li3=[]
li4=[]
li5=[]
li6=[]
li7=[]
li8=[]
li9=[]
L=[li1,li2,li3,li4,li5,li6,li7,li8,li9]
concats=[]
```

```python
# (카테고리, 분기)조합별로 새로 객체 생성하여 매칭되는 증감값 컬럼4개 추가
for i,k in zip(카테고리,L):
    for j,p in zip(분기,range(4)):
        k.append(total_addQc[(total_addQc['카테고리']==i)&(total_addQc['분기']==j)])
        k[p]['PER증감']=PER.loc[i][j]
        k[p]['PBR증감']=PBR.loc[i][j]
        k[p]['ROE증감']=ROE.loc[i][j]
        k[p]['EBIT증감']=EBIT.loc[i][j]
        print(len(total_addQc[(total_addQc['카테고리']==i)&(total_addQc['분기']==j)]))
```

```python
# 모두 병합
for i in range(9):
    for j in range(4):
        concats.append(L[i][j])
        print(len(L[i][j]))
TotalFinal=pd.concat(concats,axis=0)
TotalFinal=TotalFinal.sort_values(by='Unnamed: 0')
TotalFinal.head()
```
