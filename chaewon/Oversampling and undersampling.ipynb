```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, roc_auc_score, f1_score
```

```python
data = pd.read_excel('마스터.xlsx')
```

```python
data.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, inplace=True)
```

```python
data['시각화지수'] = data.좋아요수 + (data.이미지수*10) + (100 - data.비디오수*10)
```

```python
data.columns
```

```python
selected_features = ['목표금액', 
       '글업데이트수', '리워드종류수',
       '와디즈팔로워수', '과거프로젝트수', '과거성공프로젝트수', '앵콜펀딩여부',
       '음절수', '단어수', '문장수', 'Strong', 'Under', '가독성', '가독성2', '제목단어수',
       '디자인소품', '반려동물', '뷰티', '스포츠·모빌리티', '여행·레저',
       '테크·가전', '패션·잡화', '푸드', '홈리빙', 'sns팔로워수', 'Fri', 'Mon', 'Sat', 'Sun',
       'Thu', 'Tue', 'Wed', '음절수_제목', '단어수_제목', '문장수_제목', '가독성_제목', '가독성_제목.1',
       '펀딩기간', '일펀딩금액', '일글수', '문장당강조', '문장당밑줄', '문장지수', '시각화지수', '달성률']
```

```python
selected_data = data[selected_features]
```

```python
target_list = []

for i in range(len(selected_data)):
    if selected_data.iloc[i]['달성률'] < 100:
        target_list.append(0)
    elif selected_data.iloc[i]['달성률'] < 850:
        target_list.append(1)
    else:
        target_list.append(2)
```

```python
selected_data['target'] = target_list
```

```python
selected_data
```

```python
# 타겟별 분포 보기
selected_data.pivot_table(index='target', aggfunc='count', values='달성률')
```

## Random forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

X = selected_data.drop(['달성률', 'target'], axis=1)
y = selected_data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

rf_clf = RandomForestClassifier()

rf_params = {
    'n_estimators': [3000, 4000, 5000],
    'class_weight': [None, 'balanced'],
    'max_depth': [5, 7, 9, 11]
}

grid_rf = GridSearchCV(rf_clf, param_grid=rf_params, cv=5)

grid_rf.fit(X_train, y_train)
```

```python
grid_rf.best_estimator_.score(X_test, y_test)
```

```python
confusion_matrix(y_test, grid_rf.best_estimator_.predict(X_test))
```

## SMOTE (oversampling)

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=0)

# 오버샘플링된 train_data 만들기
X_train_over, y_train_over = smote.fit_resample(X_train, y_train)
```

```python
# 오버샘플링 전 타겟 분포
y_train.value_counts()
```

```python
# 오버샘플링 후 타겟 분포
pd.Series(y_train_over).value_counts()
```

```python
grid_rf.fit(X_train_over, y_train_over)
```

```python
grid_rf.best_estimator_.score(X_test, y_test)
```

```python
confusion_matrix(y_test, grid_rf.best_estimator_.predict(X_test))
```

## ADASYN (oversampling)

```python
from imblearn.over_sampling import ADASYN

ada = ADASYN(random_state=0)

X_train_ada, y_train_ada = ada.fit_resample(X_train, y_train)
```

```python
pd.Series(y_train_ada).value_counts()
```

```python
grid_rf.fit(X_train_ada, y_train_ada)

print(grid_rf.best_estimator_.score(X_test, y_test))
print(confusion_matrix(y_test, grid_rf.best_estimator_.predict(X_test)))
```

## TomekLinks (undersampling)

```python
from imblearn.under_sampling import TomekLinks

tl = TomekLinks(sampling_strategy='auto')

X_train_tl, y_train_tl = tl.fit_resample(X_train, y_train)
```

```python
pd.Series(y_train_tl).value_counts()
```

```python
grid_rf.fit(X_train_tl, y_train_tl)

print(grid_rf.best_estimator_.score(X_test, y_test))
print(confusion_matrix(y_test, grid_rf.best_estimator_.predict(X_test)))
```
