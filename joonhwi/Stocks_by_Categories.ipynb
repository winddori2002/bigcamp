```python
import pandas as pd
from pandas import DataFrame
import numpy as np
```

```python
filename = '(1)StocksbyCategories.xlsx'
data = pd.read_excel(filename, sheet_name=None)
```

```python
TechElec, Beauty, Food, Deco, Fashion, Homeliving, Leisure, Sports, Pet = data.values()
#data.keys()
```

```python
def mk_avg(df):
    df_clmn = df.columns
    new_ls = []
    for _ in range(1,len(df_clmn)):
        ls = list(df[df_clmn[_]])
        avg_value = (sum(ls)-max(ls)-min(ls)) / (len(ls)-2)
        new_ls.append(avg_value)
    new_row = pd.DataFrame([new_ls], columns = df_clmn[1:])
    result = pd.concat([df,new_row])
    return result
```

```python
new_TechElec = mk_avg(TechElec)
new_Beauty = mk_avg(Beauty)
new_Food = mk_avg(Food)
new_Deco = mk_avg(Deco)
new_Fashion = mk_avg(Fashion)
new_Homeliving = mk_avg(Homeliving)
new_Leisure = mk_avg(Leisure)
new_Sports = mk_avg(Sports)
new_Pet = mk_avg(Pet)
```
