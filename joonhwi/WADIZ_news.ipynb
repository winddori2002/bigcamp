```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import datetime
```

```python
data = pd.read_excel('와디즈_뉴스개수.xlsx')
```

```python
all_date = pd.date_range(start = '2018-01-01',end = '2019-08-14')
all_date_df = pd.DataFrame(all_date)
all_date_df.columns = ['날짜']
all_date_new_df = pd.merge(all_date_df, data, how='outer')
```

```python
all_date_new_df['기사제목'] = all_date_new_df['기사제목'].fillna(0)
```

```python
all_date_new_df
```

```python
new_date = all_date_new_df['날짜']
new_sum = all_date_new_df['기사제목']
```

```python
new_ls = []
for dt in range(0, len(new_date)):
    sum = new_sum[dt]
    n = 7 if len(new_date)-dt > 7 else len(new_date) - dt
    for _ in range(dt+1, dt+n):
        result = new_date[_] - new_date[dt]
        if(result.days <= 6):
            sum = sum+new_sum[_]
    new_ls.append(sum)
all_date_new_df['일주일개수'] = new_ls
```
